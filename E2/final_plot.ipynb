```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from matplotlib.ticker import NullFormatter
plt.rcParams['font.size'] = 14
# from IPython.display import display_html 
```

```python
dir_name = "My_data_results_FINALE_2hidden_fin/"
Number_of_files = 5
best = "Adam_20_one_hot_NO-Spin"
file_name = dir_name + best

cols=["epochs", "energy", "A_S", "A_T", "score", "entropy", "log_likelihood"]
df = pd.DataFrame(columns=cols)

for i in range(1, Number_of_files+1):
    read_result = pd.read_csv(file_name+"_"+str(i)+".dat", delimiter=" ", names=cols, header=0)
    df = df.append(read_result, ignore_index=True)
```

```python
df
```

```python
res_mean = df.groupby(["epochs"]).mean()    # mean values
res_std = df.groupby(["epochs"]).std()      # error bars
```

```python
# res_mean
```

```python
plt.figure(figsize=(8,8))
plt.tight_layout()
color = "cornflowerblue"
err_color = "royalblue"
var = "energy"
# plot of the AAI index
plt.plot(df["epochs"].unique(), res_mean[var], label='AAI', ls="--", marker="o", color=color)
plt.errorbar(df["epochs"].unique(), res_mean[var], yerr=res_std[var], fmt='.', capsize=1, color=err_color)
plt.xlabel("# epochs")
plt.ylabel("Energy")
plt.legend()
plt.grid()
# plt.savefig("energy.png", dpi=300)
plt.show()
```
